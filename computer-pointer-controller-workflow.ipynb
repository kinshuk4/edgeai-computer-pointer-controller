## Step 1.1: Submit to an Edge Compute Node with an Intel CPU
In the cell below, write a script to submit a job to an <a 
    href="https://software.intel.com/en-us/iot/hardware/iei-tank-dev-kit-core">IEI 
    Tank* 870-Q170</a> edge node with an <a 
    href="https://ark.intel.com/products/88186/Intel-Core-i5-6500TE-Processor-6M-Cache-up-to-3-30-GHz-">Intel Core™ i5-6500TE processor</a>. The inference workload should run on the CPU.

```python
def get_job_arguments(scenario, device_type, device_name, max_people):
    video_path="/data/resources/{}.mp4".format(scenario)
    queue_param_path="/data/queue_param/{}.npy".format(scenario)
    output_path="/output/results/{}/{}".format(scenario, device_type.lower())
    #MODEL=$1 DEVICE=$2 VIDEO=$3 QUEUE=$4 OUTPUT=$5 PEOPLE=$6
    job_arguments="{} {} {} {} {} {}".format(model_path, device_name, video_path, 
                                             queue_param_path, output_path, max_people)
    
    return job_arguments

def submit_job(scenario, device_type, device_name, model_path, edge_node, max_people):
    
    job_arguments = get_job_arguments(scenario, device_type, device_name, max_people)
    job_name="{}_{}".format(scenario, device_type)
    job = !qsub queue_job.sh -d . -l nodes=1:{edge_node} -F "{job_arguments}" -N {job_name}
    
    return job
```

```python
scenario="manufacturing"
model_path="/data/models/intel/person-detection-retail-0013/FP16/person-detection-retail-0013"
max_people=2
```

```python
device_name="CPU"
device_type="CPU"
edge_node="tank-870:i5-6500te"
```

```python
cpu_job_id=submit_job(scenario, device_type, device_name, model_path, edge_node, max_people)
print(cpu_job_id[0])
```

#### Check Job Status

## Step 2: Assess Performance

Run the cells below to compare the performance across all 4 devices. The following timings for the model are being compared across all 4 devices:

- Model Loading Time
- Average Inference Time
- FPS

```python
import matplotlib.pyplot as plt

device_list=['cpu', 'gpu', 'fpga', 'vpu']
inference_time=[]
fps=[]
model_load_time=[]

for device in device_list:
    with open('results/manufacturing/'+device+'/stats.txt', 'r') as f:
        inference_time.append(float(f.readline().split("\n")[0]))
        fps.append(float(f.readline().split("\n")[0]))
        model_load_time.append(float(f.readline().split("\n")[0]))
```

```python
plt.bar(device_list, inference_time)
plt.xlabel("Device Used")
plt.ylabel("Total Inference Time in Seconds")
plt.show()
```

```python
plt.bar(device_list, fps)
plt.xlabel("Device Used")
plt.ylabel("Frames per Second")
plt.show()
```

```python
plt.bar(device_list, model_load_time)
plt.xlabel("Device Used")
plt.ylabel("Model Loading Time in Seconds")
plt.show()
```
